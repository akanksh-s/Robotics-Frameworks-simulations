## 🖼️ Task 1: Basic Image Operations
We will load an image, convert it to different color spaces, and apply thresholding.

```python
# 1. Reading the image
image = cv2.imread('robots.png')
if image is None:
    print("Error: 'robots.png' not found. Please upload it in the cell above.")
else:
    print("Original Image (BGR to RGB for display):")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_RGB)
    plt.axis('off')
    plt.show()

    # 2. Converting to grayscale
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print("\nGrayscale Image:")
    cv2_imshow(image_gray)

    # 3. Binary Thresholding
    # Param 175 is the threshold value. Pixels > 175 become 255 (white).
    thresh_val, image_binary = cv2.threshold(image_gray, 175, 255, cv2.THRESH_BINARY)
    print(f"\nBinary Image (Threshold: {thresh_val}):")
    cv2_imshow(image_binary)
```

## 🌫️ Task 2: Smoothing and Blurring
Blurring is often used as a preprocessing step to reduce noise or detail level.

```python
image_smoothing = cv2.imread('robots.png')
kernel_sizes = [(3, 3), (9, 9), (15, 15)]

plt.figure(figsize=(15, 5))
print("Comparing Mean (Average) Blurring:")
for i, (x, y) in enumerate(kernel_sizes):
    blurred = cv2.blur(image_smoothing, (x, y))
    plt.subplot(1, 3, i+1)
    plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
    plt.title(f"Kernel {x}x{y}")
    plt.axis('off')
plt.show()

plt.figure(figsize=(15, 5))
print("\nComparing Gaussian Blurring:")
for i, (x, y) in enumerate(kernel_sizes):
    blurred = cv2.GaussianBlur(image_smoothing, (x, y), 0)
    plt.subplot(1, 3, i+1)
    plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
    plt.title(f"Kernel {x}x{y}")
    plt.axis('off')
plt.show()

plt.figure(figsize=(15, 5))
print("\nComparing Median Blurring:")
for i, k in enumerate([3, 9, 15]):
    blurred = cv2.medianBlur(image_smoothing, k)
    plt.subplot(1, 3, i+1)
    plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
    plt.title(f"K-size {k}")
    plt.axis('off')
plt.show()
```

## 📐 Task 3: Canny Edge Detection
The Canny algorithm uses a multi-stage process to detect a wide range of edges in images.

```python
image_canny = cv2.imread('robots.png')
# Canny(image, minThreshold, maxThreshold)
edges_canny = cv2.Canny(image_canny, 50, 200)

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image_canny, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(edges_canny, cmap='gray')
plt.title('Edge Image (Canny)')
plt.axis('off')
plt.show()
```

## 🌍 Task 4: Perspective Transformation

**Note:** In local Python, we use `cv2.setMouseCallback` to pick points. In Google Colab, we use pre-defined coordinates or pick them manually from a display.

We want to "straighten" the piece of paper in the image.

```python
paper = cv2.imread('paper.jpeg')
if paper is None:
    print("Error: 'paper.jpeg' not found.")
else:
    # Coordinates of the 4 corners of the paper in the original image
    # (upper left, upper right, lower left, lower right)
    # These coordinates are calibrated for the provided 'paper.jpeg'
    points_before = np.float32([[92, 24], [220, 73], [26, 98], [172, 174]])

    # Dimensions for the output image (we use original width/height)
    width, height = paper.shape[1], paper.shape[0]
    points_after = np.float32([[0, 0], [width, 0], [0, height], [width, height]])

    # Mark the selected points in the original image for visualization
    paper_marked = paper.copy()
    for pt in points_before:
        cv2.circle(paper_marked, (int(pt[0]), int(pt[1])), 5, (0, 255, 0), -1)

    # Calculate the Projection Matrix M
    M = cv2.getPerspectiveTransform(points_before, points_after)

    # Apply the transformation
    paper_transformed = cv2.warpPerspective(paper, M, (width, height))

    # Display Results
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(paper_marked, cv2.COLOR_BGR2RGB))
    plt.title('Original with Points')
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(cv2.cvtColor(paper_transformed, cv2.COLOR_BGR2RGB))
    plt.title('Perspective Corrected')
    plt.axis('off')
    plt.show()
```
